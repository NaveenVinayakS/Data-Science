<b>Types of categorical variable</b>
<p>1. Nominal data and encoding</p>
<p>2. Ordinal data and encoding</p>

In Ordinal data, while encoding, one should retain the information regarding the <b>order</b> in which the category is provided. Like in the above example the highest degree a person possesses, gives vital information about his qualification. The degree is an important feature to decide whether a person is suitable for a post or not.

While encoding Nominal data, we have to consider the presence or absence of a feature. In such a case, no notion of <b>order</b> is present. For example, the city a person lives in. For the data, it is important to retain where a person lives. Here, We do not have any <b>order</b> or sequence. It is equal if a person lives in Delhi or Bangalore.

<b>Feature Scaling</b>

Values in the features will be in madnitude(values) and units(kg,cm,meter).

<b>Role of Scaling</b> is mostly important in algorithms that are distance based and require Euclidean Distance. Random Forest is a tree-based model and hence does not require feature scaling.

<b>Rule of thumb I follow here is any algorithm that computes distance or assumes normality, scale your features!!!</b>

some algorithm defanetly need to perform feature scaling eg:(Knn) 
<b>k-nearest neighbors</b> with an Euclidean distance measure is sensitive to magnitudes and hence should be scaled for all features to weigh in equally.

Scaling is critical, while performing <b>Principal Component Analysis(PCA)</b>. PCA tries to get the features with maximum variance and the variance is high for high magnitude features. This skews the PCA towards high magnitude features.

We can speed up <b>gradient descent</b> by scaling. This is because θ will descend quickly on small ranges and slowly on large ranges, and so will oscillate inefficiently down to the optimum when the variables are very uneven.(very helpful in <b>Linear Regression</b> for gradient descent to obtain Global minimum)

<b>Tree based models</b> are not distance based models and can handle varying ranges of features. Hence, Scaling is not required while modelling trees.

Algorithms like <b>Linear Discriminant Analysis(LDA), Naive Bayes</b> are by design equipped to handle this and gives weights to the features accordingly. Performing a features scaling in these algorithms may not have much effect.


Standardisation and Mean Normalization can be used for algorithms that assumes zero centric data like Principal Component Analysis(PCA).

Euclidean distance is used in these algorithm so we need to perform scaling.
